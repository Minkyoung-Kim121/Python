```python
x = np.arange(-5,5,0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2

plt.figure(1) # >>> 이미 생성된 그래프 창 1
plt.plot(x,y1)

plt.figure(2) # >>> "" 창 2
plt.plot(x,y2)

plt.figure(1)
plt.plot(x,y2)

plt.figure(2)
plt.clf()
plt.plot(x,y1)

plt.show()
```

```python
# 그래프 창을 하위 그래프 영역으로 나누기
# plt.subplot(m,n,p)

x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

# 2 x 2 행렬 그래프에서 p 위치 지정
# 1
plt.subplot(2,2,1)
plt.plot(x,y1)

# 2
plt.subplot(2,2,2)
plt.plot(x,y2)

# 3
plt.subplot(2,2,3)
plt.plot(x,y3)

# 4
plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
# 그래프 확대해서 보기
x = np.linspace(-4,4,100)
# >>> -4 ~ 4 까지 100개 값 생성
y1 = x**3
y2 = 10*x**2 -2

plt.plot(x,y1,x,y2)
plt.show()
# >>> 겹쳐진 영역이 잘 안 보임
```

```python
# 겹쳐진 영역 확대해보기
# xlim()
# ylim()
# >>> 그래프의 축 범위 지정 또는 가져오기 함수
# >>> 위의 두 함수를 이용하기 위해서는 그래프를 출력한 후에 실행해야한다.
plt.plot(x,y1,x,y2)
plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
```

```python
# 그래프 꾸미기 (=그꾸)
# fmt = '[color][line_style][marker]'
# b : blue
# g : green
# r : red
# c : cyan
# m : magenta
# y : yellow
# k : black
# w : white

# solid line : 실선
# dashed line : 파선
# dotted line : 점선
# dash-dot line : 파선&점선

# o : 원 모양
# ^,v,>,< : 삼각형 위,아래,오,왼 방향
# s : square
# p : pentagon(오각형)
# h, H : hexagon(육각형)
# * : star
# + : plus
# x,X : 채워진 x
# D,d : diamond
```

```python
# 지역별 인구밀도
import numpy as np

city = ['서울','인천','대전','대구','울산','부산','광주']

# latitude & longitude
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

# marker size
size = np.array(pop_den) * 0.2
colors = ['r','g','b','c','m','k','y']

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('경도')
plt.ylabel('위도')
plt.title('인구밀도')
plt.grid()

for x,y, name in zip(lon, lat, city):
    plt.text(x,y,name)

plt.show()
```

```python
# var gragh
# 막대 그래프
# 데이터 비교 시 사용
# plt.bar(x, data[,width=width_f, color=colors, tick_label=tick_labels, 
#         align='conter'(기본) 또는 'edge', label=labels])

member_IDs = ['m1','m2','m3','m4']
before_ex = [27,35,40,33]
after_ex = [30,38,42,37]

import matplotlib.pyplot as plt
import numpy as np

n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.show()
```

```python
# add option
plt.bar(index, before_ex, tick_label = member_IDs)
plt.show()
```

```python
# bar color
colors = ['y','c','b','pink']
plt.bar(index, before_ex, color = colors, tick_label=member_IDs)
plt.show()
```

```python
# width
plt.bar(index, before_ex, tick_label=member_IDs, width=0.5)
plt.show()
```

```python
# 가로로 표현
# barh()
colors = ['pink', 'skyblue','c','gray']
plt.barh(index, before_ex, color = colors, tick_label=member_IDs)
plt.show()
```

```python
barWidth = 0.4
# >>> 두 막대 그래프 들어가게 여유롭게 구성
plt.bar(index, before_ex, color='c', align='edge',width=barWidth, label='before')
plt.bar(index + barWidth, after_ex, color='m', align='edge', width=barWidth, label='after')

plt.xticks(index + barWidth, member_IDs)
plt.legend()
plt.xlabel('회원ID')
plt.ylabel('운동횟수')
plt.title('운동 시작 전과 후의 복근 변화')
plt.show()
```

```python
# histogram
# 히스토그램
# 데이터의 분포

# 변량 : 자료를 측정해 숫자로 표시한 것
# 계급 : 변량을 정해진 간격으로 나눈 구간
# 계급의 간격 : 계급을 나눈 크기
# 도수 : 나눠진 계급에 속하는 변량의 수
# 도수분포표 : 계급에 도수를 표시한 표

# plt.hist(x,[,bins=bins_n or 'auto'])

import matplotlib.pyplot as plt
math = [76,82,84,83,90,86,85,92,72,71,100,87,81,76,94,78,81,60,79,
       69,74,87,82,68,79]
plt.hist(math)
# >>> hist()는 기본으로 변량을 10개의 계급으로 나누어서 표시.
```

```python
# bins option
plt.hist(math, bins=8)
plt.show()
# >>> 8개의 계급으로 나눔
```

```python
plt.hist(math, bins=8)
plt.xlabel('시험점수')
plt.ylabel('도수')
plt.title('수학 시험의 히스토그램')
plt.grid()
plt.show()
```

```python
# pie graph
# 파이 그래프
# 파이 조각 처럼 생겨서 >>> 전체 데이터에서 각 항목이 차지한 비율 보고싶을 때 사용
# plt.figure(figsize=(w,h))

fruit = ['사과','바나나','오렌지','딸기','망고']
result = [7,6,3,2,2]

import matplotlib.pyplot as plt
plt.pie(result)
plt.show()
```

```python
# 비율 맞추기
plt.figure(figsize=(5,5))
plt.pie(result)
plt.show()
```

```python
# 라벨 추가
plt.figure(figsize=(5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%')
plt.show()
```

```python
# 옵션 지정
plt.figure(figsize=(5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
# 사과 강조하기
explode_value = (0.1,0,0,0,0)

plt.figure(figsize=(5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90, counterclock=False,
       explode=explode_value, shadow=True)
plt.show()
```
