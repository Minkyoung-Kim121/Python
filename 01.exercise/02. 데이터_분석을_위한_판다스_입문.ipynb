```python
# 박스 그래프
# 이산형 변수(명확히 구분되는 값) 와 연속형 변수(셀 수 없는 범위 값)를 같이 사용하는 그래프
# boxplot()
# lables 인자값 : 성별을 구분하기 위한 이름 추가
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
             labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
# 다변량 그래프 : 변수 3개 이상 사용한 그래프
# 점의 크기 다르게, 투명도 조절 등으로 그래프 표현

# 산점도 그래프에 새 변수 추가
# 0,1 정수 사용해서 구분해야 >>> 각 문자열을 정수로 변환하는 함수 생성 필요.
# 여성 >>> 0
# 남성 >>> 1
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

# 반환 값 0,1을 데이터 프레임에 추가
# sex_color 열 새로 생성해서 추가
# apply()
tips['sex_color'] = tips['sex'].apply(recode_sex)

# 테이블당 인원수 추가 >>> 점 크기로 표현
# s : 점 size
# c : 점 color
# alpha : 투명도
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'],
             y=tips['tip'],
             s=tips['size']*10,
             c=tips['sex_color'],
             alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# 누락값의 개수 반환
import numpy as np
def count_missing(vec):
    null_vec = pd.isnull(vec)
    null_count = np.sum(null_vec)
    return null_count
```

```python
cmis_col = titanic.apply(count_missing)
print(cmis_col)
# age : 177
# embarked : 2
# deck : 688
# embark_town : 2
```

```python
# 누락값의 비율 계산 함수
def prop_missing(vec):
    num = count_missing(vec)
    dem = vec.size
    return num / dem
```

```python
pmis_col = titanic.apply(prop_missing)
print(pmis_col)
```

```python
# 누락값이 아닌 데이터의 비율
def prop_complete(vec):
    return 1 - prop_missing(vec)
```

```python
# 행 방향 누락값 처리
cmis_row = titanic.apply(count_missing, axis=1)
pmis_row = titanic.apply(prop_missing, axis=1)
pcom_row = titanic.apply(prop_complete, axis=1)

print(cmis_row.head())
print("="*50)
print(pmis_row.head())
print("="*50)
print(pcom_row.head())
```

```python
ebola.index = ebola['Date']
new_idx = pd.date_range(ebola.index.min(), ebola.index.max())
print(new_idx)
```

```python
# reversed() : 인덱스를 반대로 뒤집는 메서드
new_idx = reversed(new_idx)
```

```python
# reindex(a) : a를 새로운 인덱스로 지정
ebola = ebola.reindex(new_idx)
print(ebola.head().iloc[:, :5])
```

```python
print(ebola.tail().iloc[:, :5])
```

```python
# last_valid_index() : 가장 오래된 데이터 가져오기
# first_valid_index() : 가장 최근의 데이터 가져오기
last_valid = ebola.apply(pd.Series.last_valid_index)
print(last_valid)
```

```python
first_valid = ebola.apply(pd.Series.first_valid_index)
print(first_valid)
```

```python
earliest_date = ebola.index.min()
print(earliest_date)
```

```python
shift_values = last_valid - earliest_date
print(shift_values)
```

```python
ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
```

```python
ebola_shift = pd.DataFrame(ebola_dict)
print(ebola_shift.tail())
```

```python
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date','Day'], axis=1)
print(ebola_shift.tail())
```
